### Problema 3

```python
# Ingresar el primer número
num1 = int(input("Ingrese el primer número: "))

# Ingresar el segundo número
num2 = int(input("Ingrese el segundo número: "))

# Sumar los dos números
resultado = num1 + num2

# Mostrar el resultado y el mensaje correspondiente
if resultado < 200:
    print("El resultado es menor a 200")
elif resultado < 250:
    print("El resultado es mayor a 200, pero menor a 250")
else:
    print("El resultado es mayor a 250")
```

### Problema 4

```python
def evaluar_usuario(edad, gusta_musica_urbana):
    # Viendo si el usuario es mayor de edad y además, gusta la música urbana
    if edad >= 18 and gusta_musica_urbana:
        mensaje = "Eres mayor de edad y te gusta la música urbana."
    elif edad >= 18 and not gusta_musica_urbana:
        mensaje = "Eres mayor de edad, pero no te gusta la música urbana."
    elif edad < 18 and gusta_musica_urbana:
        mensaje = "Te gusta  la música urbana, pero eres menor de edad."
    else:
        mensaje = "No te gusta la música urbana y eres menor de edad."

    return mensaje

# Pedirle al usuario su edad y si le gusta la música urbana
edad = int(input("¿Cuál es tu edad? "))
gusta_musica_urbana = input("¿Te gusta la música urbana? (Sí/No): ").lower() == "sí"

# Resultado
print(evaluar_usuario(edad, gusta_musica_urbana))
```
